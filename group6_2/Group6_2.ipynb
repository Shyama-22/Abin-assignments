```python
#question 1
from Bio import Entrez, SeqIO
from io import StringIO

# Replace with your email address (required by NCBI)
Entrez.email = "shyama21492@iiitd.ac.in"

s_accession_number = "NM_001297986"

def download_sequence():
    try:
        # Fetch the sequence from NCBI in FASTA format
        handle = Entrez.efetch(db="nucleotide", id=s_accession_number, rettype="fasta", retmode="text")
        sequence = handle.read()
        handle.close()
        return sequence
    except Exception as e:
        raise ValueError(f"Failed to download sequence due to: {e}")
def count_kmers(sequence, k):
    # Generate k-mers using zip and list comprehension
    kmers = [''.join(chars) for chars in zip(*[sequence[i:] for i in range(k)])]

    kmers_counts = {}
    for kmer in kmers:
        kmers_counts[kmer] = kmers_counts.get(kmer, 0) + 1

    return kmers_counts

def count_kmers_from_string(sequence_string, k):
    # Convert the raw sequence string into a file-like object
    sequence_file_like = StringIO(sequence_string)
    records = list(SeqIO.parse(sequence_file_like, "fasta"))

    # Extract sequences from records using list comprehension
    sequences = [str(record.seq) for record in records]

    # Use list comprehension to generate kmer counts for each sequence
    kmers_lists = [count_kmers(seq, k) for seq in sequences]

    total_kmers_count = {}
    for kmers_counts in kmers_lists:
        for kmer, count in kmers_counts.items():
            total_kmers_count[kmer] = total_kmers_count.get(kmer, 0) + count

    return total_kmers_count


def main():
    while True:
        print("\nMenu:")
        print("1. Count k-mers from sequence")
        print("2. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            try:
                sequence_string = download_sequence()
                k = int(input("Please enter the value for k: "))

                if k < 1:
                    raise ValueError("k must be a positive integer.")

                result = count_kmers_from_string(sequence_string, k)
                print(result)
            except ValueError as ve:
                print(f"Error: {ve}")
            except Exception as e:
                print(f"An unexpected error occurred: {str(e)}")
        elif choice == '2':
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please choose again.")

if __name__ == "__main__":
    main()
```

```python
#question 3:
from Bio import Entrez, SeqIO

# Replace with your email address (required by NCBI)
Entrez.email = "shivam21489@iiitd.ac.in"

def download_sequence(accession):
    try:
        # Fetch the sequence from NCBI
        handle = Entrez.efetch(db="nucleotide", id=accession, rettype="gb", retmode="text")
        record = SeqIO.read(handle, "genbank")
        handle.close()

        # Store the sequence in a string variable
        sequence = str(record.seq)
        return sequence

    except Exception as e:
        print(f"An error occurred while downloading the sequence: {str(e)}")
        return None

def reconstruct_alignment(s, t, n, m, dp):  # function which calculates the reconstruct the alignment
    aligned_s, aligned_t = "", ""  # Initialize aligned sequences as empty strings
    i, j = n, m  # Initialize indices for sequence s and t to their lengths

    # Start a loop to reconstruct the alignment
    while i > 0 or j > 0:
        # Check if we can match or mismatch a character
        if i > 0 and j > 0 and (s[i - 1] == t[j - 1] or dp[i][j] == dp[i - 1][j - 1] + 1):
            # Characters at positions i-1 in s and j-1 in t match or have been substituted

            aligned_s = s[i - 1] + aligned_s
            aligned_t = t[j - 1] + aligned_t
            i -= 1  # Move to the previous positions in both s and t
            j -= 1
        # Check if we need to insert a gap in s
        elif i > 0 and dp[i][j] == dp[i - 1][j] + 1:

            aligned_s = s[i - 1] + aligned_s
            aligned_t = "-" + aligned_t
            i -= 1  # Move to the previous position in s
        else:
           # Check if we need to insert a gap in t

            aligned_s = "-" + aligned_s
            aligned_t = t[j - 1] + aligned_t
            j -= 1  # Move to the previous position in t


# reconstructs the alignment by considering matches, substitutions, insertions, and deletions

    return dp[n][m], aligned_s, aligned_t  # Return the edit distance and the aligned sequences


def global_alignment(s_accession, t_accession):
    s = download_sequence(s_accession)
    t = download_sequence(t_accession)
    #print(len(s))
    #print(len(t))
    print(f"length of the nucleotide sequence s: {len(s)}")
    print(f"length of the nucleotide sequence t: {len(t)}")
    if s is None or t is None:
        return None, None, None

    n, m = len(s), len(t)
    # initializes a 2D list dp with dimensions (n + 1) x (m + 1) and fills it with zeros
    dp = []
    for i in range(n + 1):
      row = []
      for j in range(m + 1):
          row.append(0)
      dp.append(row)

    for i in range(1, n + 1):
        for j in range(1, m + 1):
           match = dp[i - 1][j - 1] if s[i - 1] == t[j - 1] else dp[i - 1][j - 1] + 1
           delete = dp[i - 1][j] + 1
           insert = dp[i][j - 1] + 1
           dp[i][j] = min(match, delete, insert)


    return reconstruct_alignment(s, t, n, m, dp)

if __name__ == "__main__":
# Define the accession numbers for the two sequences
    s_accession_number = "NM_001301717"  # accession number for s
    t_accession_number = "NM_001001755"  # accession number for t

# Call the global_alignment function to find the alignment between the two sequences
    result = global_alignment(s_accession_number, t_accession_number)

# Check if the alignment was successful (result is not None)
    if result is not None:

        distance, s_prime, t_prime = result
# Print the edit distance and the aligned sequences
        print(f"Distance: {distance}")
        print(f"s_prime: {s_prime}")
        print(f"t_prime: {t_prime}")

# If the alignment failed, print an error message
    else:
        print("Failed to download one or both sequences.")
```
